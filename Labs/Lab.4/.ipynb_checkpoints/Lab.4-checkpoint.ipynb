1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self,max_val):
        self.max_val = max_val
        self.val=0
    def increment(self):
        if self.val < self.max_val:
            self.val +=1
        else:
            print('Error: cannot exceed max value')
    def reset(self):
        self.val=0
    def ret_val(self):
        return self.val
    
```

```python
count = Counter(5)

count.increment()
count.increment()
print(count.ret_val())

count.increment()
count.increment()
count.increment()
count.increment()

count.reset()
print(count.ret_val())
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
class Shape:
    def area(self):
        raise NotImplementedError
    def perimeter(self):
        raise NotImplementedError
    def parameter(self, n=16):
        raise NotImplementedError    

class Rectangle(Shape):
    def __init__(self,x,y,width,height):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__height = height
        
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def width(self):
        return self.__width
    def height(self):
        return self.__height
    
    def area(self):
        return self.__width* self.__height
    def perimeter(self):
        return 2*(self.__width+ self.__height)
    def parameter(self, n=16):
        pts = []
        for i in range(n):
            t = i / n
            pts.append((self.__x + t * self.__width, self.__y))
            pts.append((self.__x + self.__width, self.__y + t * self.__height))
            pts.append((self.__x + (1 - t) * self.__width, self.__y + self.__height))
            pts.append((self.__x, self.__y + (1 - t) * self.__height))
        return pts[:n]
    
class Circle(Shape):
    def __init__(self,x,y,radius):
        self.__x = x
        self.__y = y
        self.__radius=radius
        
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def radius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    def perimeter(self):
        return 2* math.pi* self.__radius
    def parameter(self, n=16):
        pts = []
        for i in range(n):
            theta = 2 * math.pi * i / n
            pts.append((self.__x + self.__radius * math.cos(theta),
                        self.__y + self.__radius * math.sin(theta)))
        return pts


class Triangle(Shape):
    def __init__(self,s1,s2,s3):
        self.__s1 = s1
        self.__s2 = s2
        self.__s3 = s3
        
        
    def s1(self):
        return self.__s1
    def s2(self):
        return self.__s2
    def s3(self):
        return self.__s3
    
    def area(self):
        x1,y1 = self.__s1
        x2,y2 = self.__s2
        x3,y3 = self.__s3
        return abs (x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2
    
    def perimeter(self):
        def d(a,b):
            return math.dist(a,b)
        return d(self.__s1,self.__s2)+d(self.__s2,self.__s3)+d(self.__s3,self.__s1)
    
    def parameter(self, n=16):
        pts = []
        edges = [(self.__s1,self.__s2),(self.__s2,self.__s3),(self.__s3,self.__s1)]
        for a,b in edges:
            for i in range(n//3):
                t = i/(n//3)
                pts.append((a[0]+t*(b[0]-a[0]), a[1]+t*(b[1]-a[1])))
        return pts[:n]
```

```python
class Shape:
    def area(self):
        raise NotImplementedError
    def perimeter(self):
        raise NotImplementedError
    def parameter(self, n=16):
        raise NotImplementedError    

class Rectangle(Shape):
    def __init__(self,x,y,width,height):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__height = height
        
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def width(self):
        return self.__width
    def height(self):
        return self.__height
    
    def area(self):
        return self.__width* self.__height
    def perimeter(self):
        return 2*(self.__width+ self.__height)
    
    def parameter(self, n=16):
        pts = []
        for i in range(n):
            t = i / n
            pts.append((self.__x + t * self.__width, self.__y))
            pts.append((self.__x + self.__width, self.__y + t * self.__height))
            pts.append((self.__x + (1 - t) * self.__width, self.__y + self.__height))
            pts.append((self.__x, self.__y + (1 - t) * self.__height))
        return pts[:n]
    def contains(self, x, y):
        return (self.__x <= x <= self.__x + self.__width and self.__y <= y <= self.__y + self.__height)
    
class Circle(Shape):
    def __init__(self,x,y,radius):
        self.__x = x
        self.__y = y
        self.__radius=radius
        
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def radius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    def perimeter(self):
        return 2* math.pi* self.__radius
    
    def parameter(self, n=16):
        pts = []
        for i in range(n):
            theta = 2 * math.pi * i / n
            pts.append((self.__x + self.__radius * math.cos(theta),
                        self.__y + self.__radius * math.sin(theta)))
        return pts
    def contains(self, x, y):
        return (x - self.__x) ** 2 + (y - self.__y) ** 2 <= self.__radius ** 2


class Triangle(Shape):
    def __init__(self,s1,s2,s3):
        self.__s1 = s1
        self.__s2 = s2
        self.__s3 = s3
         
    def s1(self):
        return self.__s1
    def s2(self):
        return self.__s2
    def s3(self):
        return self.__s3
    
    def area(self):
        x1,y1 = self.__s1
        x2,y2 = self.__s2
        x3,y3 = self.__s3
        return abs (x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2
    
    def perimeter(self):
        def d(a,b):
            return math.dist(a,b)
        return d(self.__s1,self.__s2)+d(self.__s2,self.__s3)+d(self.__s3,self.__s1)
    
    def parameter(self, n=16):
        pts = []
        edges = [(self.__s1,self.__s2),(self.__s2,self.__s3),(self.__s3,self.__s1)]
        for a,b in edges:
            for i in range(n//3):
                t = i/(n//3)
                pts.append((a[0]+t*(b[0]-a[0]), a[1]+t*(b[1]-a[1])))
        return pts[:n]
    
    def contains(self, x, y):
        A = self.area()
        t1 = Triangle(self.__s1, self.__s2, (x,y)).area()
        t2 = Triangle(self.__s2, self.__s3, (x,y)).area()
        t3 = Triangle(self.__s3, self.__s1, (x,y)).area()
        return abs(A - (t1+t2+t3)) < 1e-6
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
        
class Shape:
    def area(self):
        raise NotImplementedError
    def perimeter(self):
        raise NotImplementedError
    def parameter(self, n=16):
        raise NotImplementedError 
    def contains(self, x, y):
        raise NotImplementedError
    def overlaps(self, other):
        for x, y in self.parameter():
            if other.contains(x, y):
                return True
        for x, y in other.parameter():
            if self.contains(x, y):
                return True
        return False
    
    def paint(self, canvas, char='*'):
        for row in range(canvas.height):
            for col in range(canvas.width):
                if self.contains(col, row):
                    canvas.set_pixel(row, col, char)
                    
class CompoundShape(Shape):
    def __init__(self):
        self.__shapes = []

    def add(self, shape):
        self.__shapes.append(shape)

    def area(self):
        return sum(s.area() for s in self.__shapes)

    def perimeter(self):
        return sum(s.perimeter() for s in self.__shapes)

    def parameter(self, n=100):
        pts = []
        for s in self.__shapes:
            pts.extend(s.parameter(n))
        return pts

    def contains(self, x, y):
        return any(s.contains(x, y) for s in self.__shapes)

    def paint(self, canvas, char='*'):
        for s in self.__shapes:
            s.paint(canvas, char)

class Rectangle(Shape):
    def __init__(self,x,y,width,height):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__height = height
        
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def width(self):
        return self.__width
    def height(self):
        return self.__height
    
    def area(self):
        return self.__width* self.__height
    def perimeter(self):
        return 2*(self.__width+ self.__height)
    
    def parameter(self, n=16):
        pts = []
        for i in range(n):
            t = i / n
            pts.append((self.__x + t * self.__width, self.__y))
            pts.append((self.__x + self.__width, self.__y + t * self.__height))
            pts.append((self.__x + (1 - t) * self.__width, self.__y + self.__height))
            pts.append((self.__x, self.__y + (1 - t) * self.__height))
        return pts[:n]
    def contains(self, x, y):
        return (self.__x <= x <= self.__x + self.__width and self.__y <= y <= self.__y + self.__height)
    
class Circle(Shape):
    def __init__(self,x,y,radius):
        self.__x = x
        self.__y = y
        self.__radius=radius
        
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def radius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    def perimeter(self):
        return 2* math.pi* self.__radius
    
    def parameter(self, n=16):
        pts = []
        for i in range(n):
            theta = 2 * math.pi * i / n
            pts.append((self.__x + self.__radius * math.cos(theta),
                        self.__y + self.__radius * math.sin(theta)))
        return pts
    def contains(self, x, y):
        return (x - self.__x) ** 2 + (y - self.__y) ** 2 <= self.__radius ** 2


class Triangle(Shape):
    def __init__(self,s1,s2,s3):
        self.__s1 = s1
        self.__s2 = s2
        self.__s3 = s3
         
    def s1(self):
        return self.__s1
    def s2(self):
        return self.__s2
    def s3(self):
        return self.__s3
    
    def area(self):
        x1,y1 = self.__s1
        x2,y2 = self.__s2
        x3,y3 = self.__s3
        return abs (x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2
    
    def perimeter(self):
        def d(a,b):
            return math.dist(a,b)
        return d(self.__s1,self.__s2)+d(self.__s2,self.__s3)+d(self.__s3,self.__s1)
    
    def parameter(self, n=16):
        pts = []
        edges = [(self.__s1,self.__s2),(self.__s2,self.__s3),(self.__s3,self.__s1)]
        for a,b in edges:
            for i in range(n//3):
                t = i/(n//3)
                pts.append((a[0]+t*(b[0]-a[0]), a[1]+t*(b[1]-a[1])))
        return pts[:n]
    
    def contains(self, x, y):
        A = self.area()
        t1 = Triangle(self.__s1, self.__s2, (x,y)).area()
        t2 = Triangle(self.__s2, self.__s3, (x,y)).area()
        t3 = Triangle(self.__s3, self.__s1, (x,y)).area()
        return abs(A - (t1+t2+t3)) < 1e-6
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add(self, shape, char='*'):
        self.shapes.append((shape, char))

    def remove(self, shape):
        self.shapes = [(s, c) for (s, c) in self.shapes if s != shape]

    def clear(self):
        self.shapes.clear()

    def paint(self):
        self.canvas.data = [[' '] * self.canvas.width for _ in range(self.canvas.height)]
        for shape, char in self.shapes:
            shape.paint(self.canvas, char)

    def display(self):
        self.canvas.display()
```

```python
if __name__ == "__main__":

    drawing = RasterDrawing(60, 25)

    r = Rectangle(5, 5, 15, 8)
    c = Circle(35, 12, 6)
    t = Triangle((45,5), (55,5), (50,15))

    drawing.add(r, '#')
    drawing.add(c, '@')
    drawing.add(t, '+')

    print("Initial Drawing:\n")
    drawing.paint()
    drawing.display()

    print("\nAfter Modification:\n")

    drawing.remove(r) 

    new_circle = Circle(20, 10, 5)
    drawing.add(new_circle, '*') 
    
    drawing.paint()
    drawing.display()
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```

```python
# Test
print(repr(foo(1,"hello")))
```

```python
# Create an object and save it
ff=foo(1,"hello")
ff.save("Test.foo")
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
        
class Shape:
    def area(self):
        raise NotImplementedError
    def perimeter(self):
        raise NotImplementedError
    def parameter(self, n=16):
        raise NotImplementedError 
    def contains(self, x, y):
        raise NotImplementedError
    def overlaps(self, other):
        for x, y in self.parameter():
            if other.contains(x, y):
                return True
        for x, y in other.parameter():
            if self.contains(x, y):
                return True
        return False
    
    def paint(self, canvas, char='*'):
        for row in range(canvas.height):
            for col in range(canvas.width):
                if self.contains(col, row):
                    canvas.set_pixel(row, col, char)
                    
class CompoundShape(Shape):
    def __init__(self):
        self.__shapes = []

    def add(self, shape):
        self.__shapes.append(shape)

    def area(self):
        return sum(s.area() for s in self.__shapes)

    def perimeter(self):
        return sum(s.perimeter() for s in self.__shapes)

    def parameter(self, n=100):
        pts = []
        for s in self.__shapes:
            pts.extend(s.parameter(n))
        return pts

    def contains(self, x, y):
        return any(s.contains(x, y) for s in self.__shapes)

    def paint(self, canvas, char='*'):
        for s in self.__shapes:
            s.paint(canvas, char)

class Rectangle(Shape):
    def __init__(self,x,y,width,height):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__height = height
        
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def width(self):
        return self.__width
    def height(self):
        return self.__height
    
    def area(self):
        return self.__width* self.__height
    def perimeter(self):
        return 2*(self.__width+ self.__height)
    
    def parameter(self, n=16):
        pts = []
        for i in range(n):
            t = i / n
            pts.append((self.__x + t * self.__width, self.__y))
            pts.append((self.__x + self.__width, self.__y + t * self.__height))
            pts.append((self.__x + (1 - t) * self.__width, self.__y + self.__height))
            pts.append((self.__x, self.__y + (1 - t) * self.__height))
        return pts[:n]
    def contains(self, x, y):
        return (self.__x <= x <= self.__x + self.__width and self.__y <= y <= self.__y + self.__height)
    
class Circle(Shape):
    def __init__(self,x,y,radius):
        self.__x = x
        self.__y = y
        self.__radius=radius
        
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def radius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    def perimeter(self):
        return 2* math.pi* self.__radius
    
    def parameter(self, n=16):
        pts = []
        for i in range(n):
            theta = 2 * math.pi * i / n
            pts.append((self.__x + self.__radius * math.cos(theta),
                        self.__y + self.__radius * math.sin(theta)))
        return pts
    def contains(self, x, y):
        return (x - self.__x) ** 2 + (y - self.__y) ** 2 <= self.__radius ** 2


class Triangle(Shape):
    def __init__(self,s1,s2,s3):
        self.__s1 = s1
        self.__s2 = s2
        self.__s3 = s3
         
    def s1(self):
        return self.__s1
    def s2(self):
        return self.__s2
    def s3(self):
        return self.__s3
    
    def area(self):
        x1,y1 = self.__s1
        x2,y2 = self.__s2
        x3,y3 = self.__s3
        return abs (x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2
    
    def perimeter(self):
        def d(a,b):
            return math.dist(a,b)
        return d(self.__s1,self.__s2)+d(self.__s2,self.__s3)+d(self.__s3,self.__s1)
    
    def parameter(self, n=16):
        pts = []
        edges = [(self.__s1,self.__s2),(self.__s2,self.__s3),(self.__s3,self.__s1)]
        for a,b in edges:
            for i in range(n//3):
                t = i/(n//3)
                pts.append((a[0]+t*(b[0]-a[0]), a[1]+t*(b[1]-a[1])))
        return pts[:n]
    
    def contains(self, x, y):
        A = self.area()
        t1 = Triangle(self.__s1, self.__s2, (x,y)).area()
        t2 = Triangle(self.__s2, self.__s3, (x,y)).area()
        t3 = Triangle(self.__s3, self.__s1, (x,y)).area()
        return abs(A - (t1+t2+t3)) < 1e-6
```
